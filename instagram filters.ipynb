## 1. Documentation

 - i used the simple way to create image effect by add the overlay image the the original one 
 - i created seven filters by varying the parameters (blue,green,red,intensity,...etc)
 - after that i applied GaussianBlur filter

## 2.Script

```python
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2 

def add_overlay_with_intensity(image,blue=0,green=0,red=0,intensity=0.5,constant=0):
    h= image.shape[0]
    w=image.shape[1]
    frame = cv2.cvtColor(image, cv2.COLOR_BGR2BGRA)
    layer=np.full((h, w, 4),(blue,green,red,1), dtype='uint8')
    return cv2.addWeighted(layer, intensity, frame, 1.0,constant)

def insta_like(image,instafilter):
    if instafilter=="InKwell":
        gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        Y= cv2.cvtColor(gray, cv2.COLOR_BGR2BGRA)
        Y1=add_overlay_with_intensity(gray)
        final=cv2.GaussianBlur(Y1,(21,21),2)
    elif instafilter=="Lily":
        Y=add_overlay_with_intensity(image,blue=0,green=175,red=125,intensity=0.4,constant=0)
        final=cv2.GaussianBlur(Y,(21,21),2)
    elif instafilter=="Sutro":
        gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        Y= cv2.cvtColor(gray, cv2.COLOR_BGR2BGRA)
        Y1=add_overlay_with_intensity(Y,blue=0,green=0,red=60,intensity=0.5,constant=0)
        final=cv2.GaussianBlur(Y1,(21,21),1) 
    elif  instafilter=="Poprocket":
        y=add_overlay_with_intensity(image,blue=10,green=0,red=200,intensity=.5,constant=30)
        final=cv2.GaussianBlur(y,(21,21),1)
    elif instafilter=="Lord Kelvin":
        Y=add_overlay_with_intensity(image,blue=20,green=80,red=150,intensity=0.7,constant=0)
        final=cv2.GaussianBlur(Y,(21,21),2)
    elif instafilter=="Hefe":
        Y=add_overlay_with_intensity(image,blue=10,green=130,red=180,intensity=0.5,constant=-30)
        final=cv2.GaussianBlur(Y,(21,21),2)
    elif instafilter=="Walden":
        Y=add_overlay_with_intensity(image,blue=100,green=15,red=25,intensity=0.5,constant=-20)
        final=cv2.GaussianBlur(Y,(21,21),1)
   
    return final
```

## 3.0 Input Image

```python
img=cv2.imread('input.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

## 3.1  InKwell Filter

```python
InKwell=insta_like(img,"InKwell")
plt.imshow(cv2.cvtColor(InKwell,cv2.COLOR_BGR2RGB))
```
